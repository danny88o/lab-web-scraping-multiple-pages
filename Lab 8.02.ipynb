# Wiki Python

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
response = requests.get('https://en.wikipedia.org/wiki/Python')
response.status_code
```

```python
soup_wkp = BeautifulSoup(response.content, "html.parser")
soup_wkp
```

```python
##mw-content-text > div.mw-parser-output > ul:nth-child(7) > li:nth-child(1) > a
soup_wkp.select("#mw-content-text > div.mw-parser-output > ul > li > a")[0]["href"]
```

```python
links = []
for i in range(len(soup_wkp.select("#mw-content-text > div.mw-parser-output > ul > li > a"))):
    links.append("en.wikipedia.org" + soup_wkp.select("#mw-content-text > div.mw-parser-output > ul > li > a")[i]["href"])
links
```

# Wiki Top 10 Lang

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers')
response.status_code
```

```python
soup_wkl = BeautifulSoup(response.content, "html.parser")
soup_wkl
```

```python
#mw-content-text > div.mw-parser-output > table:nth-child(13) > tbody > tr:nth-child(1) > td:nth-child(2)
soup_wkl.select("#mw-content-text > div.mw-parser-output > table > tbody > tr > td")
```

```python
language = []
speakers = []
for i in range(0,len(soup_wkl.select("#mw-content-text > div.mw-parser-output > table > tbody > tr > td > a"))-6, 6):
    language.append(soup_wkl.select("#mw-content-text > div.mw-parser-output > table > tbody > tr > td")[i+1].get_text().split("\n")[0])
    speakers.append(soup_wkl.select("#mw-content-text > div.mw-parser-output > table > tbody > tr > td")[i+2].get_text().split("\n")[0])
```

```python
lang =  pd.DataFrame({"Language": language,
                            "Speakers in millions": speakers,
                           })
lang
```
